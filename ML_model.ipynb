```python
#importing necessary libraries to implement ML model and assess accuracy.

import numpy as np
import pandas as pd
import os
import re
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from textblob import TextBlob as TB
```

### Getting sentiment polarity scores using textblob

```python
df = pd.read_csv(r"PATH", encoding = "utf-8") #Add path to required csv file
df
```

```python
score = []
sentiment = []
label = []

for i in range(len(df)):
    fullText = df.loc[i,'cleaned_text']
    fullText = str(fullText)
    fullText = fullText.encode('utf-8')
    sentimentScore = TB(str(fullText)).sentiment.polarity
    sentimentScore = float(sentimentScore)
    score.append(sentimentScore)
    if (sentimentScore < -0.25):
        sentiment.append('Negative')
        label.append(-1)
    elif (sentimentScore > 0.25):
        sentiment.append('Positive')
        label.append(1)
    else:
        sentiment.append('Neutral')
        label.append(0)

df['textblob_Score'] = score
df['textblob_Sentiment'] = sentiment
df['textblob_Label'] = label

df
```

### Sentiment Classification using supervised machine learning algorithm

```python
data1=pd.read_csv(r"PATH", encoding = "utf-8") #Add path to required csv file
#data1
```

```python
#Divide data into training and testing set - 70:30

from sklearn.model_selection import train_test_split

X = data1['cleaned_text']
y = data1['textblob_Label']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10, train_size = .7,shuffle = True)
```

```python
# BAG OF WORDS implementation using count vector - bigram method

from sklearn.feature_extraction.text import CountVectorizer

cVec=CountVectorizer(ngram_range=(1,2))
cVec_train=cVec.fit_transform(X_train.values.astype('U'))
cVec_test = cVec.transform(X_test.values.astype('U'))
data_count = cVec.transform(data1.iloc[:,10].values.astype('U'))
```

```python
#Parameters for checking accuracy.

def classification_params(test,predic):

    from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
    
    matrix=confusion_matrix(test,predic)
    print(matrix)
    score=accuracy_score(test,predic)
    print(score)
    report=classification_report(test,predic)
    print(report)
    return matrix, score, report
```

```python
def sentiment(score):
    sentiment = []
    for i in range(len(data1)):
        sentimentScore = data1.loc[i,score]
        if (sentimentScore < -0.25):
            sentiment.append('Negative')
        elif (sentimentScore > 0.25):
            sentiment.append('Positive')
        else:
            sentiment.append('Neutral')
    return sentiment
```

#### 1. Random Forest

```python
#implement RandomForest Classifier for count vector

from sklearn.ensemble import RandomForestClassifier

randomclassifier=RandomForestClassifier(n_estimators=50,criterion='entropy')

randomclassifier.fit(cVec_train,y_train)
random_count_pred = randomclassifier.predict(cVec_test)

classification_params(y_test,random_count_pred)
```

```python
data1['Random_score'] = randomclassifier.predict(data_count)

data1['Random_sentiment'] = sentiment('Random_score')
```

#### 2. Support Vector Machine

```python
# Perform classification with SVM, kernel=linear
from sklearn import svm

classifier_linear = svm.SVC(kernel='linear')

classifier_linear.fit(cVec_train, y_train)
prediction_linear = classifier_linear.predict(cVec_test)

classification_params(y_test,prediction_linear)
```

```python
data1['svc_lin_score'] = classifier_linear.predict(data_count)

data1['svc_lin_sentiment'] = sentiment('svc_lin_score')
```

```python
# Perform classification with SVM, kernel=rbf


classifier_rbf = svm.SVC()

classifier_rbf.fit(cVec_train, y_train)
prediction_rbf = classifier_rbf.predict(cVec_test)

classification_params(y_test,prediction_rbf)
```

```python
data1['svc_rbf_score'] = classifier_rbf.predict(data_count)

data1['svc_rbf_sentiment'] = sentiment('svc_rbf_score')
```
