## Cleaning text

```python
import pandas as pd
import re

#Convert csv file into dataframe
df = pd.read_csv('path') ## add your file path here

def remove_links(text):
    return re.sub(r'http\S+', '', text) #To remove web links from tweets' text

def remove_nonalphanum(text):
    return re.sub('[^a-zA-Z0-9\s]', '', text) #To remove any characters other than alphanumerics from tweets' text

def remove_hashtags(text):
    return re.sub(r'@\w+|#\w+', '', text)  #To remove @username mentions from tweets' text

def newline_space(text):
    return re.sub(r'\s*\n\s*',' ', text)   #To remove extra spaces and new lines from tweets' text

df['new_text'] = df['text'].apply(remove_links)
df['new_text'] = df['new_text'].apply(remove_nonalphanum)
df['new_text'] = df['new_text'].apply(newline_space)
df['new_text'] = df['new_text'].str.lower()   #Converting text into lowercase

#save dataframe as csv
df.to_csv('path', index=False) ##change here too
```

## Removing stopwords from text

```python
from sklearn.feature_extraction.text import CountVectorizer

df = pd.read_csv('path') ##change path

vectorizer = CountVectorizer(stop_words='english') #Create a CountVectorizer object with English stopwords

X = vectorizer.fit_transform(df['new_text'])
feature_names = vectorizer.get_feature_names_out()

cleaned_text = [' '.join([word for word in text.split() if word.lower() not in vectorizer.get_stop_words()]) for text in df['new_text']]
df_cleaned = pd.DataFrame({'cleaned_text': cleaned_text})
df = pd.concat([df, df_cleaned], axis=1)  #Create a new DataFrame with the cleaned text

df.to_csv('path', index=False) ##change path
```
